```python
from pathlib import Path
import pandas as pd
import urllib.request

def collect_data():
    # Create data directory if it doesn't exist
    Path('data').mkdir(parents=True, exist_ok=True)
    
    # Define file path and URL
    path = Path('data/ipl_matches.csv')
    url = "https://raw.githubusercontent.com/12345k/IPL-Dataset/master/IPL/data.csv"
    
    # Download file if it doesn't exist
    if not path.is_file():
        urllib.request.urlretrieve(url, path)
        
    return pd.read_csv(path)


```

```python
match_data = collect_data()
```

```python
dropped_features = ["date", "id", "player_of_match", "win_by_runs", "win_by_wickets", "city", "season", "umpire1", "umpire2", "result"]
match_data = match_data.drop(dropped_features, axis=1)
```

```python
home_grounds = {
    "Royal Challengers Bangalore": "M Chinnaswamy Stadium",
    "Chennai Super Kings": "MA Chidambaram Stadium, Chepauk",
    "Delhi Daredevils": "Feroz Shah Kotla",
    "Kings XI Punjab": "Punjab Cricket Association Stadium, Mohali",
    "Kolkata Knight Riders": "Eden Gardens",
    "Mumbai Indians": "Wankhede Stadium",
    "Rajasthan Royals": "Sawai Mansingh Stadium",
    "Sunrisers Hyderabad": "Rajiv Gandhi International Stadium, Uppal"
}

def classify_venue(row):
    venue = row['venue']
    team1 = row['team1']
    team2 = row['team2']
    
    team1_ground = home_grounds.get(team1)
    team2_ground = home_grounds.get(team2)
    
    if venue == team1_ground:
        return team1  # return team1's name if it's their home ground
    elif venue == team2_ground:
        return team2  # return team2's name if it's their home ground
    else:
        return 'neutral ground'  # return neutral ground if neither is at their home ground

# Apply the function to each row
match_data['venue_type'] = match_data.apply(classify_venue, axis=1)
```

```python
match_data.drop("venue", axis=1, inplace=True)
```
